```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python

sns.distplot(df['fare'])
```

```python
sns.jointplot(x='age',y='fare',data=df,kind='scatter')
```

```python
sns.barplot(x='sex', y='fare', data=df)
```

```python
sns.countplot(x='class', data=df)
```

```python
sns.barplot(x='class', y='fare', data=df)
```

```python
sns.barplot(x='class', y='fare', hue='sex', data=df)
```

```python
sns.heatmap(df.corr(),cmap='coolwarm')
plt.title('Correlation Heatmap')
```

```python
sns.FacetGrid(data=df,col='sex').map(plt.hist,'age')
```

```python
sns.lmplot(x='age', y= 'survived', hue= 'class', data=df)
```
